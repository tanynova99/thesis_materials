```python
from pytube import YouTube
from youtube_transcript_api import YouTubeTranscriptApi
import pandas as pd
```

# Main architechture

```python
def ytvid_to_text(video_url):
  try:
      # Create a YouTube object
      yt = YouTube(video_url)

      # Get the highest resolution stream
      video_stream = yt.streams.get_highest_resolution()

      # Download the video
      video_stream.download()

      # Get the video ID from the URL
      video_id = video_url.split("v=")[1]
      # print(video_id)

      # Get the video transcript
      transcripts = YouTubeTranscriptApi.get_transcript(video_id)

      if transcripts:
        # Extract the transcripts as text
        subtitle_text = " ".join([entry["text"] for entry in transcripts])

        return subtitle_text

      else:
        print("Transcripts not available for this video.")

  except Exception as e:
    print(f"An error occurred: {str(e)}")
    return 'empty'
```

# Adding transcripts to csv file



```python
url = 'https://docs.google.com/spreadsheets/d/1wopAoFHAc41lcWp0__TdKSc3I6OzSbn_VnB6n-nieP0/export?format=csv'
links_raw = pd.read_csv(url)
links = links_raw['link']
links[:5]
```

```python
transcripts = links_raw.iloc[:, : 4]
list_transcripts = []

for link in links:
  # print(link)
  txt = ytvid_to_text(link)
  list_transcripts.append(txt)

print(list_transcripts)
```

```python
transcripts['transcript'] = list_transcripts
```
