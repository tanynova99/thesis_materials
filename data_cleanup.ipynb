# Reading  the experimental Data

```python
url = 'https://docs.google.com/spreadsheets/d/1ax-nLtcB9JWc4BMsxlPg8eadkUwOjzqI_EDtVftknIo/export?format=csv'
df = pd.read_csv(url)
df.head(5)
```

# Cleanup

```python
df = df.rename(columns={"What's your native language? Please use one word in English, do not capitalize it. If you are a native speaker, write 'english'.": "l1",
                        "What's your English level?":"l2"})
```

```python
df['Score'] = df['Score'].str.replace(" / 242", "")
```

```python
df = df.drop('Timestamp', axis=1)
```

```python
df['l1'] = df['l1'].apply(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
df['l1'] = df['l1'].str.strip()
```

```python
df.head(5)
```

```python
def return_assigned_number(word, l_dict):
    for key, values in l_dict.items():
      if word in values:
        return key
    return None

l1_dict = {
    int(0): {"english"},
    int(1): {"afrikaans", "spanish", "italian", "portuguese"},
    int(2): {"armenian", "cantonese", "chinese", "dutch", "persian", "swahili", "kiswahili", "turkish", "maltese"},
    int(3): {"arabic", "bulgarian", "greek", "german", "norwegian", "polish", "russian", "serbian", "finnish", "french", "czech", "swedish", "ukrainian", "kurdish", "hebrew"}
}

l2_dict = {
    "A1": {"A1 Elementary"},
    "A2": {"A2 Pre Intermediate"},
    "B1": {"B1 Intermediate"},
    "B2": {"B2 Upper Intermediate"},
    "C1": {"C1 Advanced"},
    "C2": {"C2 Proficient"},
    "NS": {"I am a native speaker"}
}

df['l2'] = df['l2'].apply(lambda x: return_assigned_number(x, l2_dict))
df['l1'] = df['l1'].apply(lambda x: return_assigned_number(x, l1_dict))
```

```python
df
```

# Removal of repeated columns and unneeded questions

```python
url = 'https://docs.google.com/spreadsheets/d/1X0th6M9-hpe7RN0DIP3XEjNHj1nsYccNUPkRDEI43Ts/export?format=csv'
questions = pd.read_csv(url)
```

```python
df_merged = df
df_merged["Story"] = np.nan

for column in list(df.columns)[4:]:
  for question in questions['question']:
    if  question in column:
      if question != column:
        df_merged[question] = df_merged[question].combine_first(df_merged[column])
        df_merged = df_merged.drop(column, axis=1)
  if '#3' in column:
    df_merged["Story"] = df_merged["Story"].combine_first(df_merged[column])
    df_merged = df_merged.drop(column, axis=1)
```

```python
to_drop = ["My ski instructor believes it will snow in the mountains tomorrow evening.",
           "I make breakfast every day.",
           "Three days ago, I spend a very interesting evening at my friend’s house.",
           "The cat has just caught a mouse.",
           "There were no sausages left when I came back. The dog had eaten them.",
           "I ____ in London next year, doing the same old job. (be)",
           "I'm going now but I ____ you at the same time tomorrow. (see)",
           "If I don't have too much work to do this year, I think I ____ the novels I am planning to write by the end of the year. (finish)",
           "Everybody ____ money. It is very important. (need)",
           "I usually ____ the children a story before they go to bed. (tell)",
           "I usually ____ as a secretary. (work)",
           "It was the first time the actor ____ such bad reviews. (receive)",
           "First she ____ the advertisement, then she called the company. (read)",
           "They ____ the programme twice on TV already. (broadcast)",
           "They ____ Bob Matthews an Oscar. It was on the news just now. (award)",
           "We ____ two big Macs before we went home. (eat)"]



for stimul in to_drop:
  df_merged = df_merged.drop(columns=stimul, axis=1)
```
